```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data = data.drop(columns=['Unnamed: 0'])
data.head()
```

```python
data.info()
```

```python
data.describe().transpose()
```

```python
x = data.copy()
x = x.drop(columns = ['class_type'])
y = data['class_type']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.head()
```

SVM

```python
from sklearn.svm import SVC
import pickle
svc_model  = SVC(kernel='poly')
svc_model.fit(x_train, y_train)
print("**********poly***********")
prediction = svc_model .predict(x_test)
# check the accuracy on the training set
print("training",svc_model.score(x_train, y_train))
print("validation",svc_model.score(x_test, y_test))
filename = 'svc_model_4.sav'
pickle.dump(svc_model, open(filename, 'wb'))
```

```python
from sklearn.svm import SVC
import pickle
  
# Building a Support Vector Machine on train data
svc_model = SVC(C= .1, kernel='linear', gamma= 1)
svc_model.fit(x_train, y_train)
print("**********linear***********")
prediction = svc_model .predict(x_test)
# check the accuracy on the training set
print("training",svc_model.score(x_train, y_train))
print("validation",svc_model.score(x_test, y_test))

svc_model = SVC(kernel='rbf')
svc_model.fit(x_train, y_train)
print("**********rbf***********")
prediction = svc_model .predict(x_test)
# check the accuracy on the training set
print("training",svc_model.score(x_train, y_train))
print("validation",svc_model.score(x_test, y_test))


svc_model  = SVC(kernel='poly')
svc_model.fit(x_train, y_train)
print("**********poly***********")
prediction = svc_model .predict(x_test)
# check the accuracy on the training set
print("training",svc_model.score(x_train, y_train))
print("validation",svc_model.score(x_test, y_test))
filename = 'svc_model_4.sav'
pickle.dump(svc_model, open(filename, 'wb'))

svc_model  = SVC(kernel='sigmoid')
svc_model.fit(x_train, y_train)
print("**********sigmoid***********")
prediction = svc_model .predict(x_test)
# check the accuracy on the training set
print("training",svc_model.score(x_train, y_train))
print("validation",svc_model.score(x_test, y_test))
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train, y_train)
prediction_results = knn.predict(x_test)
filename = 'knn_model.sav'
pickle.dump(knn, open(filename, 'wb'))
print(prediction_results)
print("training",knn.score(x_train, y_train))
print("validation",knn.score(x_test, y_test))
```

Decision Tree

```python
from sklearn import tree
dtc = tree.DecisionTreeClassifier()

dtc.fit(x_train, y_train)
prediction_results = dtc.predict(x_test)
print(prediction_results)
print("training",dtc.score(x_train, y_train))
print("validation",dtc.score(x_test, y_test))
```

```python
classes = {"pointer":0,"left_click":1,"right_click":2,"hold_drag":3,"left_dbl_click":4,"scrolldown":5,"scrollup":6,"right_arrow":7,"left_arrow":8}
one_hot_y_train = []
for label in y_train:
  row = np.zeros(9)
  row[classes[label]] = 1
  one_hot_y_train.append(row)

one_hot_y_test = []
for label in y_test:
  row = np.zeros(9)
  row[classes[label]] = 1
  one_hot_y_test.append(row)
```

```python
one_hot_y_train = np.array(one_hot_y_train)
one_hot_y_test = np.array(one_hot_y_test)
```

DL Model

```python
from keras.models import Sequential 
from keras.layers import Dense, Dropout
from keras.layers import Convolution2D, MaxPool2D
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
```

```python
model = Sequential()

model.add(Dense(63, input_shape = (63,),activation = 'relu'))
model.add(Dropout(0.1))

model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.25))

model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(9, activation= 'softmax'))

model.compile(loss= 'categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
history = model.fit(x_train,one_hot_y_train, batch_size = 16 , epochs = 100, validation_data=(x_test,one_hot_y_test))
```

```python
filename = 'DL_model_4.sav'
pickle.dump(model, open(filename, 'wb'))
```
